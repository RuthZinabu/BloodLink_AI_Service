```python
import pandas as pd
import numpy as np

# Generate dates
dates = pd.date_range(start="2023-01-01", end="2024-12-31")

data = []

for date in dates:

    base_demand = np.random.randint(30, 60)

    # Weekend demand slightly lower
    if date.weekday() >= 5:
        base_demand -= np.random.randint(5,10)

    # Random emergency spike
    emergency = np.random.choice([0,10,15], p=[0.8,0.15,0.05])

    total = base_demand + emergency

    data.append({
        "date": date,
        "O+": int(total * 0.37),
        "A+": int(total * 0.34),
        "B+": int(total * 0.10),
        "AB+": int(total * 0.04)
    })

df = pd.DataFrame(data)

df.head()
```

```python
df.to_csv("../data/blood_demand_data.csv", index=False)
```
